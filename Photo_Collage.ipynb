# Photo Collage using opencv

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
images = cv2.imread("image/center.jpeg")
images = cv2.cvtColor(images,cv2.COLOR_BGR2RGB)  #to grayscale --- cv2.cvtColor(images,cv2.COLOR_BGR2GRAY and plt.imshow(images,cmap="gray)
plt.imshow(images)
plt.show()
```

```python
#operating system module --- it allows us to access different files from our system when given path
import os
```

```python
image_name = os.listdir("image")
print(image_name)
```

```python
for i in image_name:
    if i.endswith(('.jpg', '.jpeg')):
        file_path = "image/" + i
        img = cv2.imread(file_path)
        img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
        plt.imshow(img)
        plt.show()
```

```python
#image collage logic
red_img = np.zeros((50,50,3),dtype = int)
red_img[:,:,0]=200
plt.imshow(red_img)
plt.show()

green_img = np.zeros((50,50,3),dtype = int)
green_img[: , : , 1] = 250
plt.imshow(green_img)
plt.show()
```

```python
red_green = np.hstack((red_img,green_img))
plt.imshow(red_green)
plt.show()
```

```python
green_red = red_green[: , ::-1,:]
plt.imshow(green_red)
plt.show()
```

```python
collage = np.vstack((red_green,green_red))
plt.imshow(collage)
plt.show()
```

```python
animals=[]
for i in image_name:
    if i.endswith(('.jpg', '.jpeg')):
        file_path = "image/" + i
        img = cv2.imread(file_path)
        img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
        img= cv2.resize(img,(100,100))
        animals.append(img)
```

```python
border = np.zeros((2,200,3),dtype='uint8')
animals_12 = np.hstack((animals[3],animals[4]))
animals_34 = np.hstack((animals[0],animals[1]))
fin_collage = np.vstack((animals_12,border,animals_34))
plt.imshow(fin_collage)
plt.show()
```
