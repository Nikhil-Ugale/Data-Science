# Data Visualization

It is the graphical represenation of data using the charts, maps, graphs in order to understand trends, outliers and patterns in data.

```python
import matplotlib.pyplot as plt   #pyplot is class in matplotlib library and plt is an object which has certain methods
```

```python
import numpy as np
```

```python
x = np.array([1,2,3,4,5])
y= x**2 + 3
print(x,y)
```

```python
plt.style.use('seaborn')
```

plotting line plot of x and y using plot method

```python
plt.plot(x,y)
plt.show()
```

```python
#multiple line plot on same plot

plt.plot(x,x**2+3,label='Microsoft')
plt.plot(x,x**2,label='Amazon')
plt.title("Line plot for stock market")
plt.xlabel('time stamp')
plt.ylabel('stock price')
plt.legend() #we have to call legend method else we cannot label the line plots
plt.show()
```

```python
#Scatter plot
plt.style.available?
```

```python
plt.style.use("dark_background")
plt.scatter(x,x**2+3,label='Microsoft')
plt.scatter(x,x**2,label='Amazon')
plt.title("scatter plot for stock market")
plt.xlabel('time stamp')
plt.ylabel('stock price')
plt.legend() #we have to call legend method else we cannot label the line plots
plt.show()
```
