```python
import cv2
import matplotlib.pyplot as plt
plt.style.use("seaborn")
```

```python
img = cv2.imread('egmage.jpg')
print(img)
```

```python
print(img.shape)
```

```python
plt.imshow(img)
plt.show()
```

Note:- Opencv will read image in BGR format but we havee to display it in RGB 

```python
#Converting it in RGB format
rgb_image = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(rgb_image)
plt.show()
```

```python
rgb_image.shape
```

```python
plt.imshow(rgb_image)
plt.axis('off')
plt.show()
```

```python
crop_image  = rgb_image[10:190,40:210,:]   #consists of all the 3 levels of color space
plt.imshow(crop_image)
plt.show()
```

```python
#resize
crop_image = cv2.resize(rgb_image,(150,150))
plt.imshow(crop_image)
plt.show()
```
